# Import Libraries

```python
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

# start (get all main category urls)

```python
# www.drugs.com
# www.drugs.com/alpha/a.html

main_url = "https://www.drugs.com"
result = requests.get(main_url)
soup = BeautifulSoup(result.text, "html.parser")

#get all main categories urls (start with a, b, c, ... z)
main_categories_urls_list = [main_url + x.get('href') for x in soup.find_all("a", href=re.compile("/alpha"))]
main_categories_urls_list = main_categories_urls_list[1:27]

print(str(len(main_categories_urls_list)) + " fetched categories URLs")
print("Some examples:")
print(main_categories_urls_list[0])
print("...")
print(main_categories_urls_list[25])
```

## Get all subcategory urls

```python
# www.drugs.com/alpha/a.html
# https://www.drugs.com/alpha/ab.html

sub_category_urls_list = []

print ("Running...")

no_of_urls = len(main_categories_urls_list)

for index, url in enumerate(main_categories_urls_list) :
    result = requests.get(url)
    result.text[:1000]
    soup = BeautifulSoup(result.text, "html.parser")
    div = soup.find(class_= "paging-list-wrap")
    subcategory_url = [main_url + x.get('href') for x in div.find_all("a", href=re.compile("/alpha/"))]
    #sub_category_urls_list.append(subcategory_url)
    sub_category_urls_list.extend(subcategory_url)
    print(str(np.round(((index+1)/no_of_urls)*100,2)) +'% Completed' )

print("Done")
```

```python
#https://www.drugs.com/alpha/ab.html
#https://www.drugs.com/mtm/abacavir.html

drug_urls_list = []
drug_name_list = []

print ("Running...")

no_of_sub_urls = len(sub_category_urls_list)

for index, sub_url in enumerate(sub_category_urls_list):
    result = requests.get(sub_url)
    result.text[:5000]
    soup = BeautifulSoup(result.text, "html.parser")
    div = soup.find(class_= "ddc-list-column-2")
    
    if div is None:
        div = soup.find(class_= "ddc-list-unstyled")

    for x in div.find_all("a", href=re.compile("/")):
        drug_urls_list.append(main_url + x.get('href'))
    
        drug_name_list.append(x.string)
    
        if len(drug_urls_list) != len(drug_name_list):
            break    
    if len(drug_urls_list) != len(drug_name_list):
        break    

    print(str(np.round(((index+1)/no_of_sub_urls)*100,2)) +'% Completed' )
    
print("Done")
```

## Get all comments url for each drug

```python
#https://www.drugs.com/mtm/abacavir.html
#https://www.drugs.com/comments/abacavir/

drug_comment_urls_list = []
second_drug_name_list = []

print ("Running...")

no_of_drug_urls = len(drug_urls_list)

for index, drug_url in enumerate(drug_urls_list):
    result = requests.get(drug_url)
    result.text[:5000]
    soup = BeautifulSoup(result.text, "html.parser")
    div = soup.find(class_= "more-resources-list more-resources-list-general")
    #print(div)
    if div is None:
        #continue 
        div = soup.find(class_= "user-reviews-title mgb-1") #added 12/april/2020
        if div is None:
            continue
            
    drug_url = [main_url + x.get('href') for x in div.find_all("a", href=re.compile("/comments/"))]
    
    if drug_url == []:
        continue
    
    drug_comment_urls_list.extend(drug_url)
    second_drug_name_list.append(drug_name_list[index])
    
    if len(drug_comment_urls_list) != len(second_drug_name_list):
        break
    
    print(str(np.round(((index+1)/no_of_drug_urls)*100,2)) +'% Completed' )
    
print("Done")
```

```python
df = pd.DataFrame({'second_drug_name':second_drug_name_list, 'drug_comment':drug_comment_urls_list})
df.to_csv("second_drug_name.csv")
```

```python
##get table review links

#https://www.drugs.com/comments/abacavir/
#https://www.drugs.com/comments/abacavir/for-hiv-infection.html

main_url = "https://www.drugs.com"

condition_list = []
score_list = []
review_url_list = []
third_drug_name_list = []

no_of_drug_comment_urls = len(unique_drug_comment_urls_list)

print ("Running...")

for index, drug_comment_url in enumerate(unique_drug_comment_urls_list):
    result = requests.get(drug_comment_url)
    result.text[:1000]
    soup = BeautifulSoup(result.text, "html.parser")
    for h in soup.find_all('table', class_='data-list ddc-table-sortable'):
        for condition in h.find_all("th", scope = "row", class_=""):
            condition_list.append(condition.text)
            
            third_drug_name_list.append(unique_second_drug_name_list[index])
        
        #for review_url in soup.find_all("a", href=re.compile("for")):
        #    review_url_list.append(main_url + review_url.get("href"))
        for t_soup in soup.find_all('td', class_='text-nowrap'):
            for b in t_soup.find_all("a", href=re.compile("for-")):
                
                only_comment = b.get("href")
                if "comment" not in only_comment:
                    continue
                
                review_url_list.append(main_url + b.get("href"))
                
        len_condition = len(condition_list)
        len_review = len(review_url_list)
        if len_condition != len_review:
            break
    if len_condition != len_review:
        print(drug_comment_url)
        break
        
    print(str(np.round(((index+1)/no_of_drug_comment_urls)*100,2)) +'% Completed' )
    
print("Done")
```

```python
### print all pages items for each condition of each drug
#sample_url_list = ['https://www.drugs.com/comments/amoxicillin/for-dental-abscess.html', 'https://www.drugs.com/comments/acetaminophen/for-pain.html' ]

main_url = "https://www.drugs.com"

date_list = []
rating_list = []
comment_list = []
fourth_drug_name = []
total_condition_list = []

no_of_drug_comment_urls = len(review_url_list)

print ("Running...")

for index, review_url in enumerate(review_url_list):
    page_list = []
    result = requests.get(review_url)
    result.text[:1000]
    soup = BeautifulSoup(result.text, "html.parser")
    
    ### get 2nd pages onwards's url
    for h in soup.find_all('div', class_='contentBox'):
        for b in h.find_all('ul', class_="ddc-paging ddc-paging-result ddc-paging-comments"):
            #get next page urls
            b = b.find_all('li', class_="")

            for i in b:
                page = [main_url + x.get('href') for x in i.find_all("a", href=re.compile("/"))]
                page_list.append(page)

    page_list = [item for elem in page_list for item in elem] #convert list of list to list

    ### get all items in one page
    for h in soup.find_all('div', class_='contentBox'):
        for b in h.find_all('div', class_="ddc-comment"):
            #get date
            date = b.find('span',{'class':'comment-date text-color-muted'})
            if date == None:
                date = ""
                date_list.append(date)
            else:
                date_list.append(date.text)

            #get rating
            rating = b.find('div',{'class':'rating-score'})
            if rating == None:
                rating = "0"
                rating_list.append(rating)
            else:
                rating_list.append(rating.text)

            #get comment
            comment = b.find('p',{'class':'ddc-comment-content'})
            comment = comment.find('span')
            
            if comment == None:
                comment = ""
                comment_list.append(comment)
                fourth_drug_name.append(third_drug_name_list[index])
                total_condition_list.append(condition_list[index])
            else:
                comment = comment.text.replace(",", "")
                comment_list.append(comment)
                fourth_drug_name.append(third_drug_name_list[index])
                total_condition_list.append(condition_list[index])

    ### print 2nd page forward

    for page in page_list:
        result = requests.get(page)
        result.text[:1000]
        soup = BeautifulSoup(result.text, "html.parser")

        for h in soup.find_all('div', class_='contentBox'):
            for b in h.find_all('div', class_="ddc-comment"):
                #get date
                date = b.find('span',{'class':'comment-date text-color-muted'})
                if date == None:
                    date = ""
                    date_list.append(date)
                else:
                    date_list.append(date.text)

                #get rating
                rating = b.find('div',{'class':'rating-score'})
                if rating == None:
                    rating = "0"
                    rating_list.append(rating)
                else:
                    rating_list.append(rating.text)

                #get comment
                comment = b.find('p',{'class':'ddc-comment-content'})
                comment = comment.find('span')
                
                if comment == None:
                    comment = ""
                    comment_list.append(comment)
                    fourth_drug_name.append(third_drug_name_list[index])
                    total_condition_list.append(condition_list[index])
                else:
                    comment = comment.text.replace(",", "")
                    comment_list.append(comment)
                    fourth_drug_name.append(third_drug_name_list[index])
                    total_condition_list.append(condition_list[index])
                    
    print(str(np.round(((index+1)/no_of_drug_comment_urls)*100,2)) +'% Completed' )
    
print("Done")
```

```python
df = pd.DataFrame({'date':date_list, 'drug_name': fourth_drug_name,'condition':total_condition_list,'rating':rating_list, 'comment':comment_list})
df.to_csv("14_4_2020_web_scraping.csv")
```
